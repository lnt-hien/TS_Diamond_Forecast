```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt



from darts import TimeSeries
from darts.models import NBEATSModel
from darts.models import NHiTSModel
from darts.models import NLinearModel
from darts.models import DLinearModel


from darts.dataprocessing.transformers import Scaler, MissingValuesFiller
from darts.metrics import mape, r2_score
from darts.metrics import mae
from darts.datasets import EnergyDataset
```

```python
df = pd.read_csv('df_diamond.csv',index_col=[0])[::-1]
df.reset_index(inplace=True, drop=True)
df.date = pd.to_datetime(df.date)
df.set_index('date',inplace=True)
df
```

```python
def display_forecast(pred_series, ts_transformed, forecast_type, start_date=None):
    plt.figure(figsize=(8, 5))
    if start_date:
        ts_transformed = ts_transformed.drop_before(start_date)
    ts_transformed.univariate_component(0).plot(label="actual")
    pred_series.plot(label=("historic " + forecast_type + " forecasts"))
    plt.title(
        "R2: {}".format(r2_score(ts_transformed.univariate_component(0), pred_series))
    )
    plt.legend()
```

```python
df_day_avg = df.groupby(df.index.astype(str).str.split(" ").str[0]).mean().reset_index()
filler = MissingValuesFiller()
scaler = Scaler()
series = scaler.fit_transform(
    filler.transform(
        TimeSeries.from_dataframe(
            df_day_avg, "date", ["diamond price"]
        )
    )
).astype(np.float32)

series.plot()
plt.title("diamond price")
```

## NBEATS

```python
train, val = series.split_after(pd.Timestamp("20200101"))
```

```python
nbeats_gen = NBEATSModel(
    input_chunk_length=30,
    output_chunk_length=7,
    generic_architecture=True,
    num_blocks=3,
    num_layers=4,
    layer_widths=512,
    n_epochs=100,
    nr_epochs_val_period=1,
    batch_size=500,
    model_name="nbeats_generic_run",
)
nbeats_gen.to_cpu()

nbeats_int = NBEATSModel(
    input_chunk_length=30,
    output_chunk_length=7,
    generic_architecture=False,
    num_blocks=3,
    num_layers=4,
    layer_widths=512,
    n_epochs=100,
    nr_epochs_val_period=1,
    batch_size=500,
    model_name="nbeats_interpretable_run",
)

nhits = NHiTSModel(
    input_chunk_length=30, 
    output_chunk_length=7,
    n_epochs=100,
    nr_epochs_val_period=1,
    batch_size=500,
    model_name="nbeats_interpretable_run",)

nhits.to_cpu()

nlinear = NLinearModel(
    input_chunk_length=30, 
    output_chunk_length=7,
    n_epochs=100,
    nr_epochs_val_period=1,
    batch_size=500,)

dlinear = DLinearModel(
    input_chunk_length=30, 
    output_chunk_length=7,
    n_epochs=100,
    nr_epochs_val_period=1,
    batch_size=500,)
```

```python
nbeats_gen.fit(series=train, val_series=val, verbose=True)
nbeats_int.fit(series=train, val_series=val, verbose=True)
nhits.fit(train,val_series=val, verbose=True)
nlinear.fit(train,val_series=val, verbose=True)
dlinear.fit(train,val_series=val, verbose=True)
```

```python
nbeats_ge_pred_series = nbeats_gen.historical_forecasts(
    series,
    start=pd.Timestamp("20200101"),
    forecast_horizon=7,
    stride=5,
    retrain=False,
    verbose=True,
)

nbeats_int_pred_series = nbeats_int.historical_forecasts(
    series,
    start=pd.Timestamp("20200101"),
    forecast_horizon=7,
    stride=5,
    retrain=False,
    verbose=True,
)

nhits_pred_series = nhits.historical_forecasts(
    series,
    start=pd.Timestamp("20200101"),
    forecast_horizon=7,
    stride=5,
    retrain=False,
    verbose=True,
)
nlinear_pred_series = nlinear.historical_forecasts(
    series,
    start=pd.Timestamp("20200101"),
    forecast_horizon=7,
    stride=5,
    retrain=False,
    verbose=True,
)
dlinear_pred_series = dlinear.historical_forecasts(
    series,
    start=pd.Timestamp("20200101"),
    forecast_horizon=7,
    stride=5,
    retrain=False,
    verbose=True,
)
```

```python
nbeats_gen.save('model/nbeats_generic.pt')
nbeats_int.save('model/nbeats_intepret.pt')
nhits.save('model/nhits.pt')
nlinear.save('model/nlinear.pt')
dlinear.save('model/dlinear.pt')
```

```python
display_forecast(nbeats_ge_pred_series, series, "7 day", start_date=pd.Timestamp("20200101"))
display_forecast(nbeats_int_pred_series, series, "7 day", start_date=pd.Timestamp("20200101"))
display_forecast(nhits_pred_series, series, "7 day", start_date=pd.Timestamp("20200101"))
display_forecast(nlinear_pred_series, series, "7 day", start_date=pd.Timestamp("20200101"))
display_forecast(dlinear_pred_series, series, "7 day", start_date=pd.Timestamp("20200101"))
```
